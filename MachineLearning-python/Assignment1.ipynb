Create a function which creates an n√ón array with (i,j)-entry equal to i+j.

```python
def sumIJMatrix(n = 1):
    matrix = np.zeros((n, n))
    for i in range(n):
         matrix[i] = np.arange(0+i,n+i) 
    return matrix

matrix_Q6 = sumIJMatrix(5).astype(int)
print(matrix_Q6)
```

#### Part2 - Q3

What is the class distribution? (i.e. how many instances of malignant (encoded 0) and how many benign (encoded 1)?)
This function should return a Series named target of length 2 with integer values and index = ['malignant', 'benign']

```python
def classDistribution(target,p_index = []):
    class_Dis = pd.Series(data = [0]*len(p_index),index = p_index)
    for x in target:
        if x==0:
            class_Dis['malignant'] += 1
        elif x==1:
            class_Dis['benign'] += 1
    return class_Dis

classDistribution(cancer['target'],p_index = ['malignant', 'benign'])
```
